# Imports

```python
# Imports
import pandas as pd
import numpy as np

# Plotting Imports
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# Preprocessing Imports
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder

# Modelling imports
from sklearn.cluster import AgglomerativeClustering
from sklearn_extra.cluster import KMedoids

# Hyperparameter imports
from sklearn.neighbors import NearestNeighbors

# Pipeline Imports
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline

# Metric Imports
import gower
from sklearn.metrics import silhouette_score
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import linkage
from scipy.spatial.distance import squareform
from pandas.api.types import CategoricalDtype
```

# Importing and Data Cleaning

```python
# Importing data
student_data = pd.read_csv('student_data_unsupervised_learning.csv')
student_data.head(10)
```

```python
# Removing rows with empty columns

student_data = student_data.dropna()
```

```python
# Renaming columns 
student_data = student_data.rename(
    columns = {student_data.columns[1]: 'commuting_minutes', 
               student_data.columns[2]: 'work_hours',
               student_data.columns[3]: 'credits', 
               student_data.columns[4]: 'labs'})

student_data.head()
```

```python
print(student_data['labs'].value_counts())
```

# Preprocessing

```python
# Preprocessing

commuting_order = ['0-15 mins', '16-30 mins', '31-45 mins', '46-60 mins', '> Hour']
work_hours_order = ['0', '1-10', '11-20', '21-30', '31-40']
labs_order = ['0', '1', '2', '3 or more']

labels = ['commuting_minutes', 'work_hours', 'labs']

for col in labels:
    student_data[col] = student_data[col].astype(str).str.strip()

student_data.head()

preprocessor = make_column_transformer(
    (OrdinalEncoder(categories = [commuting_order, work_hours_order, labs_order]), labels),
    (StandardScaler(), ['credits']),
    remainder = 'passthrough'
)

encoded_variables = preprocessor.fit_transform(student_data)
```

```python
# Visualizing variables to pick which unsupervised learning algorithm is best fit

columns = preprocessor.get_feature_names_out() # First converting preprocessed values into a dataframe for plotting
df_encoded_variables = pd.DataFrame(encoded_variables, columns = columns)

variables =  ['ordinalencoder__commuting_minutes', 'ordinalencoder__work_hours', 'ordinalencoder__labs', 'standardscaler__credits']
sns.pairplot(df_encoded_variables[variables], diag_kind = 'scatter')
plt.show() # Plot shows a good number of oval shaped data so 
```

# Creating PAM

```python
features = ['commuting_minutes', 'work_hours', 'credits', 'labs']

# Calculating GOW Matrix

# Manually ordering my data so that when dataset is put into gower_matrix, it will know that the data is ordinal
ordinal_commuting_order = CategoricalDtype(categories = commuting_order, ordered = True)
ordinal_work_hours_order = CategoricalDtype(categories = work_hours_order, ordered = True)
ordinal_lab_order = CategoricalDtype(categories = labs_order, ordered = True)

# Converting column into ordinal data type except for credit variable
student_data['commuting_minutes'] = student_data['commuting_minutes'].astype(ordinal_commuting_order)
student_data['work_hours'] = student_data['work_hours'].astype(ordinal_work_hours_order)
student_data['labs'] = student_data['labs'].astype(ordinal_lab_order)

# Manually inputting which variables are ordinal and which one isn't to bypass error
is_category = [True, True, False, True]

# Matrix for gower to see which rows of data are most similar to each other based on the chosen variables
gower_matrix = gower.gower_matrix(student_data[features], cat_features = is_category)
```

```python
# Picking best n_clusters

total_gower_cost = [] # Array for collecting all total_gower_cost per n_cluster
mappings = {} # Dictionary for collecting the n_cluster value and

n = range (1, 10)
for N in n:
    pam = KMedoids(n_clusters = N, metric = 'precomputed',method = 'pam', init = 'k-medoids++', random_state = 42) # Used k-medoids++ for good starting points for model training
    pam_model = pam.fit(gower_matrix)
    total_gower_cost.append(pam_model.inertia_)
    mappings[N] = total_gower_cost[-1]

print("Intertia Values")
for key, val in mappings.items():
    print(f'{key} : {val}')

plt.plot(n, total_gower_cost, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Intertia')
plt.title('Elbow method using Total Gower Cost')
plt.show()
```

```python
# Model creation

pam = KMedoids(n_clusters = 3, metric = 'precomputed', method = 'pam', random_state = 42)
pam_labels = pam.fit_predict(gower_matrix)
pam_silhouette = silhouette_score(gower_matrix, pam.labels_, metric = "precomputed")
print(pam_silhouette)
```

```python
print(pd.Series(pam_labels).value_counts())
```

# Creating Hierarichal Cluster

```python
# Visualizing potential number of clusters via dendogram

hac = AgglomerativeClustering(
    metric = "precomputed",
    linkage = "complete",
    distance_threshold = 0,
    n_clusters = None
)

hac_model = hac.fit(gower_matrix)

# based from sklearn: https://github.com/scikit-learn/scikit-learn/blob/70cf4a676caa2d2dad2e3f6e4478d64bcb0506f7/examples/cluster/plot_hierarchical_clustering_dendrogram.py
def plot_dendrogram(model, **kwargs):
    
    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    sch.dendrogram(linkage_matrix, **kwargs)

plot_dendrogram(hac_model)
```

```python
# Creating Hierarichal Cluster based on number of clusters suggested by dendogram

hac = AgglomerativeClustering(
    metric = "precomputed",
    linkage = "complete",
    distance_threshold = None,
    n_clusters = 9
)

hac_labels = hac.fit_predict(gower_matrix)
hac_silhouette = silhouette_score(gower_matrix, hac_labels, metric = "precomputed")
print(hac_silhouette)
```

```python
print(pd.Series(hac_labels).value_counts())
```

# Labelling Clusters

```python
student_data['group'] = pam_labels


cluster_profiles = student_data.groupby('group').agg({
    'commuting_minutes': lambda x: x.mode()[0],
    'work_hours': lambda x: x.mode()[0],
    'credits': 'median',
    'labs': lambda x: x.mode()[0]
})

print(cluster_profiles)
```
